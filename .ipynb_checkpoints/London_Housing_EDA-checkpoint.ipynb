A. Convert the datatype of 'Date' column to date-time format.

```python
housing_data.date = pd.to_datetime(housing_data.date)
housing_data.dtypes
```

B1. Add a new column "year" in the dataframe, which contains years only.

```python
housing_data['year'] = housing_data.date.dt.year
housing_data.head()
```

B2. Add a new column "month" as the second column in the dataframe, which contains month only.

```python
housing_data.insert(1, 'month', housing_data.date.dt.month)
housing_data.head()
```

C. Drop columns "month" and "year".

```python
housing_data.drop(['month','year'], axis=1, inplace=True)
```

```python
housing_data.head()
```

D. Show all the records where the 'No. of Crimes' is 0. And. how many such records are there?

```python
housing_data[housing_data.no_of_crimes == 0]
```

E. What is the maximum and minimum average price per year in England? 

```python
housing_data['year'] = housing_data.date.dt.year
df1 = housing_data[housing_data.area == 'england']
df1.head()
```

```python
df1.groupby('year').average_price.min()
```

```python
df1.groupby('year').average_price.max()
```

F. What is the maximum and minimum number of crimes recorded per area?

```python
housing_data.groupby('area').no_of_crimes.min()
```

```python
housing_data.groupby('area').no_of_crimes.max()
```

G. Show the total count of records of each area, where the average price is less than 100000.

```python
housing_data[housing_data.average_price < 100000].area.value_counts()
```
